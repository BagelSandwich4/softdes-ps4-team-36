```python
from gene_finder import *
from helpers import load_fasta_file
genome = load_fasta_file("data/NC_045512.2.fa")
len(genome)
```

```python

threshold = noncoding_orf_threshold(genome, 1500)
threshold
orfs = find_all_orfs_both_strands(genome)
len(orfs)
```

```python
candidate_orfs = []
for orf in orfs:
    if len(orf) > threshold:
        candidate_orfs.append(orf)
len(candidate_orfs)
```

```python
proteins = []
for orf in candidate_orfs:
    proteins.append(encode_amino_acids(orf))
proteins
```
